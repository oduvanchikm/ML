````python
matrix = np.zeros((8, 8), int)

for i in range(8):
  for j in range(8):
    if (i + j) % 2 == 0:
      matrix[i, j] = 0
    else:
      matrix[i, j] = 1

print(matrix)
````

2. Сбалансировано ли соотношение мужчин и женщин в наборе данных?

````python
gender_counts = data["Gender"].value_counts()
print(gender_counts)

total = gender_counts.sum()

gender_percent = (gender_counts / total) * 100
print(gender_percent)

if abs(gender_percent["Male"] - gender_percent["Female"]) < 1:
    print("\nThe ratio of men and women is balanced")
else:
    print("\nThe ratio of men and women is not balanced")
````

3. Правда ли, что люди с высоким уровнем стресса имеют более высокое кровяное давление, чем те, кто не подвержен стрессу

````python
high_stress = data[data['Stress Levels'] >= 70]
low_stress = data[data['Stress Levels'] <= 30]

high_stress_systolic = high_stress['Blood Pressure (s/d)'].str.extract(r'(\d+)/\d+').astype(float)
low_stress_systolic = low_stress['Blood Pressure (s/d)'].str.extract(r'(\d+)/\d+').astype(float)

mean_high = high_stress_systolic.mean()[0]
mean_low = low_stress_systolic.mean()[0]

print("Average systolic BP (high stress):", round(mean_high, 1))
print("Average systolic BP (low stress):", round(mean_low, 1))

if mean_high > mean_low:
    print("People with high stress levels tend to have higher blood pressure than those with low stress.")
else:
    print("People with high stress levels do not tend to have higher blood pressure than those with low stress.")
````

4. Правда ли, что уровень холестерина у мужчин и женщин отличается?

````python
male_cholesterol = data[data["Gender"] == "Male"]["Cholesterol Level (mg/dL)"]
female_cholesterol = data[data["Gender"] == "Female"]["Cholesterol Level (mg/dL)"]

mean_male = male_cholesterol.mean()
mean_female = female_cholesterol.mean()

print("Men's average cholesterol level:", mean_male)
print("Female's average cholesterol level:", mean_female)

if abs(mean_male - mean_female) <= 1:
    print("\nCholesterol levels are not different")
else:
    print("\nCholesterol levels are different")
````

5. Правда ли, что с возрастом женщины имеют свойство оценивать свои когнитивные способности хуже, чем в юности (порог юности - 30 лет, женщина в возрасте - 60 лет+)

````python
women = data[data['Gender'] == 'Female']

young_women = women[women['Age (years)'] < 30]
older_women = women[women['Age (years)'] >= 60]

mean_young = young_women['Cognitive Function'].mean()
mean_old = older_women['Cognitive Function'].mean()

print("Average cognitive function (women under 30):", round(mean_young, 2))
print("Average cognitive function (women 60+):", round(mean_old, 2))

if (mean_young - mean_old > 0):
    print("\nWomen aged 60 and above, on average, rate their cognitive abilities lower than women under 30.")
else:
    print("\nWomen aged 60 and above do not rate their cognitive abilities lower than women under 30.")
````

6. Влияет ли физическая активность на вес в возрасте более 50 лет?

````python
over_50 = data[data['Age (years)'] > 50]

low = over_50[over_50['Physical Activity Level'] == 'Low']['Weight (kg)']
moderate = over_50[over_50['Physical Activity Level'] == 'Moderate']['Weight (kg)']
high = over_50[over_50['Physical Activity Level'] == 'High']['Weight (kg)']

print("Average weight (low activity):", round(low.mean(), 2))
print("Average weight (moderate activity):", round(moderate.mean(), 2))
print("Average weight (high activity):", round(high.mean(), 2))

if low.mean() > moderate.mean() > high.mean():
    print("\nThe higher the level of physical activity in people over 50, the lower their average weight.")
elif low.mean() > high.mean() and moderate.mean() > high.mean():
    print("\nPeople with a high level of physical activity after the age of 50 tend to weigh less on average than those who are less active.")
else:
    print("\nThere is no clear relationship between physical activity level and weight in people over 50.")
````

7. Сравните минимальный вес мужчин с высоким и низким заработком

````python
men = data[data['Gender'] == 'Male']

low_income_men = men[men['Income Level'] == 'Low']
high_income_men = men[men['Income Level'] == 'High']

min_weight_low = low_income_men['Weight (kg)'].min()
min_weight_high = high_income_men['Weight (kg)'].min()

print("Minimum weight for low-income men:", min_weight_low)
print("Minimum weight for high-income men:", min_weight_high)

if min_weight_low > min_weight_high:
    print("\nHigh-income men can have lower minimum weight compared to low-income men.")
elif min_weight_low < min_weight_high:
    print("\nLow-income men can have lower minimum weight compared to high-income men.")
else:
    print("\nMinimum weight is the same for both low- and high-income men.")
````

8. Правда ли, что у людей с индексом массы тела больше 75го перцентиля встречаются хронические заболевания чаще, чем у людей с индесом массы тела в переделах от 30 до 60 перцентиля?

````python
bmi_30 = data['BMI'].quantile(0.30)
bmi_60 = data['BMI'].quantile(0.60)
bmi_75 = data['BMI'].quantile(0.75)

group_high_bmi = data[data['BMI'] > bmi_75]
group_mid_bmi = data[(data['BMI'] >= bmi_30) & (data['BMI'] <= bmi_60)]

high_bmi_with_disease = group_high_bmi['Chronic Diseases'].notna().mean()
mid_bmi_with_disease = group_mid_bmi['Chronic Diseases'].notna().mean()

print("Proportion with chronic diseases (BMI > 75th percentile):", round(high_bmi_with_disease, 2))
print("Proportion with chronic diseases (BMI 30–60th percentile):", round(mid_bmi_with_disease, 2))

if high_bmi_with_disease > mid_bmi_with_disease:
    print("\nPeople with BMI above the 75th percentile are more likely to have chronic diseases than those in the 30th–60th percentile range.")
else:
    print("\nPeople with BMI above the 75th percentile are not more likely to have chronic diseases than those in the 30th–60th percentile range.")
````

9. Правда ли, что люди с высоким уровнем дохода имеют больший уровень физической активности, чем люди с низким доходом

````python
low_income = data[data['Income Level'] == 'Low']
high_income = data[data['Income Level'] == 'High']

low_activity_dist = low_income['Physical Activity Level'].value_counts(normalize=True)
high_activity_dist = high_income['Physical Activity Level'].value_counts(normalize=True)

print("Low income - activity levels:")
print(low_activity_dist)

print("\nHigh income - activity levels:")
print(high_activity_dist)

low_high_activity = low_activity_dist.get('High', 0)
high_high_activity = high_activity_dist.get('High', 0)

if high_high_activity > low_high_activity:
    print("\nPeople with high income tend to have higher levels of physical activity than those with low income.")
else:
    print("\nHigh-income individuals are more physically active than low-income individuals.")
````

5. Постройте круговую диаграмму отображающую как ментальное здоровье связано с паттернами сна

````python
grouped = data.groupby(['Sleep Patterns', 'Mental Health Status']).size().unstack(fill_value=0)

plt.figure(figsize=(8, 8))

outer_sizes = [sum(values) for values in grouped.values]
outer_labels = grouped.index.tolist()
wedges_outer = plt.pie(
    outer_sizes,
    radius=1.0,
    labels=outer_labels,
    labeldistance=0.9,
    colors=['gold', 'lightcoral', 'lightblue'],
    wedgeprops=dict(edgecolor='w'),
    textprops={'fontsize': 9}
)

inner_sizes = grouped.stack()
inner_labels = [f"{status}" for status in grouped.stack().index.get_level_values(1)]
wedges_inner = plt.pie(
    inner_sizes,
    radius=0.8,
    labels=inner_labels,
    labeldistance=0.7,
    colors=['green', 'blue', 'violet', 'pink'],
    wedgeprops=dict(edgecolor='w'),
    textprops={'fontsize': 9}
)

plt.title('The Impact of Sleep Patterns on Mental Health', fontsize=12)
plt.tight_layout()
plt.show()
````
